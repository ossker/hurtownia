```python
import pandas as pd
import matplotlib.pyplot as plt
from database.db_managers import OracleDbManager

db = OracleDbManager()
```

```python
# Wskaźnik unikalności kierunku - na ilu uczelniach jest dany kierunek
data = db.get_data("""
SELECT 
    kierunek.nazwa AS nazwa_kierunku,
    COUNT(DISTINCT uczelnia.nazwa) AS liczba_uczelni
FROM 
    statystyki
JOIN 
    kierunek ON statystyki.kierunek_id = kierunek.id
JOIN 
    uczelnia ON statystyki.uczelnia_id = uczelnia.id
GROUP BY 
    kierunek.nazwa
ORDER BY 
    liczba_uczelni ASC
""")

unique_subjects = pd.DataFrame(data)
unique_subjects
```

```python
# Wskaźnik produkcji absolwentów względem ilości kierunków
data = db.get_data("""
SELECT 
    nazwa_uczelni,
    SUM(ilosc) AS laczna_liczba_absolwentow,
    COUNT(nazwa_kierunku) AS liczba_kierunkow,
    ROUND(SUM(ilosc) / COUNT(nazwa_kierunku), 2) AS absolwenci_na_kierunek
FROM 
    absolwenci
GROUP BY 
    nazwa_uczelni
ORDER BY 
    absolwenci_na_kierunek DESC
""")

graduate_per_subject = pd.DataFrame(data)
graduate_per_subject
```
